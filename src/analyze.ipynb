```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
input_file = '../data/raw_data/spotify_history.csv'
df_spotify = pd.read_csv(input_file)
```

```python
df_spotify.head()
```

```python
df_2024 = df_spotify[df_spotify['ts'].str.contains('2024')]
```

```python
df_2024.head(10)
```

```python
total_time_ms = df_2024['ms_played'].sum()

total_listening_hours = total_time_ms / 3600000
total_listening_minutes = total_time_ms / 60000

print(f"Total listening hours: {total_listening_hours:.2f} hours")
print(f"Total listening minutes: {total_listening_minutes:.2f} minutes")
```

```python
total_tracks = len(df_2024)
print(f"Total tracks: {total_tracks}")
```

```python
avg_song_duration_ms = total_time_ms / total_tracks
avg_song_duration_min = avg_song_duration_ms / 60000

print(f"Avg song duration: {avg_song_duration_min:.2f} minutes")
```

```python
skipped_tracks = (df_2024['skipped'] == True).sum()
skip_rate = (skipped_tracks / total_tracks) * 100
print(f"Skip rate: {skip_rate:.1f}%")
```

```python
completed_tracks = (df_2024['reason_end'] == 'trackdone').sum()

completion_rate = (completed_tracks / total_tracks) * 100

print(f"Completion Rate: {completion_rate:.1f}%")
```

```python
df_2024['platform'].unique()
```

```python
total_time_ms = df_2024['ms_played'].sum() / 3600000
total_time_mac = df_2024[df_2024['platform'] == 'mac']['ms_played'].sum() / 3600000
total_time_android = df_2024[df_2024['platform'] == 'android']['ms_played'].sum() / 3600000

print(f"Total listening hours on mac: {total_time_mac:.1f} hours")
print(f"Total listening hours on android: {total_time_android:.1f} hours")
```

```python
df_platform = df_2024.groupby('platform').agg(
    total_hours = ('ms_played', 'sum'),
).reset_index()

df_platform
```

```python
total_time_albums = df_2024.groupby('album_name').agg(
    total_hours = ('ms_played', lambda x: x.sum() / 3600000)
).reset_index().sort_values('total_hours', ascending=False)

total_time_albums.head(10)
```

```python
top_albums = total_time_albums.head(10)
top_albums
```

```python
total_time_artists = df_2024.groupby('artist_name').agg(
    total_time = ('ms_played', 'sum'),
    total_hours = ('ms_played', lambda x: x.sum() / 3600000)
).reset_index().sort_values('total_time', ascending=False)

total_time_albums.head(10)
```

```python
df_2024['ts'] = pd.to_datetime(df_2024['ts'])

monthly_stats = df_2024.groupby(df_2024['ts'].dt.to_period('M')).agg(
    total_tracks=('track_name', 'count'),
    total_minutes=('ms_played', lambda x: x.sum() / 60000)
).reset_index()

print(monthly_stats.round(2))
```
